```python
column_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']
```

## Reorder Column Names

```python
column_names_reordered = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
df = df[column_names_reordered]
```

```python
df.head()
```

## Create a checkpoint

```python
df_reason_mod = df.copy()
```

```python
df_reason_mod.head()
```

## 'Date' column

```python
type(df_reason_mod['Date'][0])
```

```python
df_reason_mod['Date'] = pd.to_datetime(df_reason_mod['Date'], format = '%d/%m/%Y')
df_reason_mod['Date']
```

```python
type(df_reason_mod['Date'][0])
```

```python
df_reason_mod.info()
```

## extract month value

```python
df_reason_mod['Date'][0].month
```

```python
list_months = []
```

```python
df_reason_mod.shape
```

```python
for i in range(df_reason_mod.shape[0]):
    list_months.append(df_reason_mod["Date"][i].month)
```

```python
list_months
```

```python
len(list_months)
```

```python
df_reason_mod['Month Values']  = list_months
```

```python
df_reason_mod.head()
```

## Extract Day of the week value

```python
df_reason_mod['Date'][0].weekday()
```

```python
def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
df_reason_mod['day of the week'] = df_reason_mod['Date'].apply(date_to_weekday)
```

```python
df_reason_mod.head()
```

## remove Date column

```python
df_reason_mod = df_reason_mod.drop(['Date'], axis = 1)
```

```python
df_reason_mod.columns
```

```python
column_names_reordered = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Month Values', 'day of the week',
       'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education', 'Children',
       'Pets', 'Absenteeism Time in Hours']
```

```python
df_reason_mod = df_reason_mod[column_names_reordered]
df_reason_mod.head()
```

## create check point

```python
df_reason_date_mod = df_reason_mod.copy()
df_reason_date_mod.head()
```

## Modifying Education
